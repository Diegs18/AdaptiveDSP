```python
N = 100
x = np.random.normal(0, 2, N)
plt.plot(x)
b = [.2, .2, .2, .2, .2]
a = [1]
#b=[1]
#a = [1, .9]
y = sig.lfilter(b, a, x)

print('Mean value of y:', y.mean())
print('Variance of y', y.var())


plt.plot(y)
```

```python
total = 0
ryy = np.zeros(N)
for m in range(0, N):
    total = 0
    for n in range(0, N-m):
        total += y[n+m]*y[n]
    ryy[m] = total/N

plt.plot(ryy)
plt.xlabel('m')
plt.ylabel('Autocorrelation')
```

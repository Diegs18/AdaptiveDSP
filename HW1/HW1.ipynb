# Problem 2.10 Solved Below:

## 100 Samples

```python
#gernate a signal of 100 samples of white noise and run it through the filter created in the problem statement
b = [1]
a = [1, -.9]
x = np.random.normal(0, 1, 100)
y = sig.lfilter(b, a, x)
```

```python
meanx = x.mean()
varx = x.var()
meany = y.mean()
vary1 = y.var()

print("The Variance of the Input:", varx)
print("The Variance of the output:", vary1)
```

```python
#build and check the matrix, autocorr function defined above
r001 = autocorr(m=0, var=vary1)
r011 = autocorr(m=-1,var=vary1)
r101 = autocorr(m=1, var=vary1)
r111 = autocorr(m=0, var=vary1)
auto_corr_mat = [[r001, r011], [r101, r111]]
auto_corr_mat
```

```python
f, psdy=sig.periodogram(y)
plt.plot(f, psdy, marker = 'o')
plt.xlabel('Frequency')
plt.ylabel('psdy')
plt.title("100 Samples")
plt.show()
```
